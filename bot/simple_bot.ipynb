```python
#load words/embeddings
embeddings_dict = {}
with open("glove.6B/glove.6B.50d.txt", 'r', encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
```

```python
#clean out digits and punctuation
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def hasPunctuation(inputString):
    return any(char in string.punctuation for char in inputString)

words = embeddings_dict.keys()
words_clean = []

for word in words:
    if not hasNumbers(word) and not hasPunctuation (word):
        words_clean.append(word)     
```

```python
#finds closest embeddings
def find_closest_embeddings(last, target, used_words):
    ret = sorted(embeddings_dict.keys(), key=lambda word: 
                  spatial.distance.cosine(embeddings_dict[word], embeddings_dict[last]+ embeddings_dict[target]))
    #ensure bot doesn't choose its last word or its target word
    for i in range (len(ret)):
        if (ret[i] != target and ret[i] not in used_words):
            return ret[i]
```

```python
#Game set up
start_words = ['months', 'spokesman', 'campaign', 'miles', 'looking', 'justice', 'rebels', 'ruling', 'buy', 'love']
#player1_target_words = random.choices (words_clean, k=10)
#player2_target_words = random.choices ([x for x in words_clean if x not in player1_target_words], k=10)
player1_target_words = ['autopsy', 'astroturf', 'coupled', 'hydrated', 'bagged', 'blasphemies', 'analyze', 'fart', 'theatre', 'crackdowns']
player2_target_words = ['inauguration', 'evaded', 'legal', 'credit', 'poke', 'secretively', 'incensed', 'examine', 'replaceable', 'tommorrow']
start_word = random.choice (start_words)
player_1_score = 0
player_2_score = 0
last_word = start_word
used_words = [start_word]
print ("start_word: " + start_word)
print ("player1 target words: ")
print (player1_target_words)
print ("player2 target words: ")
print (player2_target_words)
print ("let the games begin!")
print ()
```

```python
for turn in range (1, 100):
    print ("turn: " + str(turn))
    
    print ("last word: " + str (last_word))
    
    player_1_cur_target = random.choice (player1_target_words)
    #player_1_cur_target = player1_target_words[1]
    print ("player 1's target in mind: " + player_1_cur_target)
    
    player_1_played_word = find_closest_embeddings (last_word, player_1_cur_target, used_words)
    last_word = player_1_played_word
    used_words.append(player_1_played_word)
    
    print ("player 1's played word: " + player_1_played_word)
    if (player_1_played_word in player2_target_words):
        print ("point for player 2!")
        player2_target_words.remove (player_1_played_word)
        player_2_score +=1
    print()
    
    
    player_2_cur_target = random.choice (player2_target_words)
    #player_2_cur_target = player2_target_words[1]
    print ("player 2's target in mind: " + player_2_cur_target)
    
    player_2_played_word = find_closest_embeddings (last_word, player_2_cur_target, used_words)
    last_word = player_2_played_word
    used_words.append(player_2_played_word)
    
    print ("player 2's played word: " + player_2_played_word)
    if (player_2_played_word in player1_target_words):
        print ("point for player 1!")
        player1_target_words.remove (player_2_played_word)
        player_1_score +=1
    print()
    
    print ('player 1 score: ' + str(player_1_score))
    print ('player 2 score: ' + str(player_2_score))
    print()
    
```
